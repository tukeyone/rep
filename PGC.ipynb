# 奖励池（total_reward_fund_pgc）
```
50*50%=25
```
奖励池总共2,500,000,000（25亿）PGC，由一个智能合约管控，按既定规则每天释放一定数量的PGC。规则为每年（每365天）释放余额的20%，即第一年释放5亿PGC，第二年4亿PGC，以此类推。那么，折算成当前每天释放的PGC数量约为1369863。每天释放的PGC被分为3部分进行分配：60%作者奖励，40%点赞奖励

````python
for x in range(1,201):
    y = 25*((1-0.2)**(x-1))*0.2
    if y*100000000>1:
        print('第'+str(x)+'年释放:',y*100000000,'每天释放：',y*100000000/365)
````

# 投票能量 （voting_power）

用户投票能量（VOTING POWER），初始值 100点 。投票时，需要消耗一定的能量，界面显示1点-100点，每次投票消耗当前剩余能量的10%，根据剩余能量不同而有所不同；为防止灌水，能量低于10点将无法投票。

取消投票不消耗能量，也不退回已经消耗的能量；修改投票不退回已经消耗的能量，并且按新的投票百分比消耗更多能量。

````python
for x in range(1,101):
    y = 100*((1-0.1)**(x-1))*0.1
    if 100*((1-0.1)**(x)) >10:
        print('第'+str(x)+'次使用:'+str(y)+'点','剩余'+str(100*((1-0.1)**(x)))+'点')
````

# 文章奖励分配(reward_article)

- 该文章奖励金额 = 每日奖励池金额 · 该文章奖励权重

文章底部所显示的收益为该文章奖励金额的预估值，该预估值的60%为作者奖励，另40%为点赞奖励。此预估值与这篇文章获得的PGC（锁定PGC+实名待领取PGC）的累计点赞数相关。

- 该文章奖励权重 = 该文章的点赞者累计贡献/所有文章的所有点赞者贡献
    - 该文章的点赞者累计贡献 = 该文章的所有点赞者贡献的累计之和
        - 该点赞者贡献（v-shares） = 该点赞者所有PGC · 该点赞者所有PGC + 2 · S  · 该点赞者所有PGC

每篇文章的奖励权重初始值为0。文章奖励权重与点赞者累计贡献PGC为非线性关系。文章奖励权重是该文章参与奖励竞争的唯一因素。

评论所获得的点赞，不会影响文章的收益计算，目前也不影响热门排序的竞争。评论可以看作是独立的文章。

- **给文章点赞的所有点赞者贡献之和**

作者奖励金额 = 该文章奖励金额 · 60% 

# 点赞奖励分配(reward_vote)

- 单一点赞者实际奖励金额 = 单一点赞者期望奖励金额 · 反向竞价比例 

    - 单一点赞者期望奖励金额 = 该文章奖励金额 · 40% · 该点赞者的点赞权重 · 位置因子
 
        - 该点赞者的点赞权重 = 该点赞者贡献/该文章所有点赞者贡献

            - 该点赞者贡献（v-shares） = 该点赞者所有PGC · 该点赞者所有PGC + 2 · S  · 该点赞者所有PGC

（1）该点赞者贡献（v-shares） = 该点赞者所有PGC · 该点赞者所有PGC + 2 · S  · 该点赞者所有PGC，其中s为固定参数。

（2）为鼓励挖掘优质内容，越早点赞的人，获得的奖励越多。在同一篇文章的点赞收益分配中，收益随着点赞的 *1/次序* 按指数级下降。

（3）同时为防止用户恶意点赞，设置了反向竞价比例，反向竞价比例计算公式为：如果当前时间 > 文章发表时间 + 30分钟，则为100%；否则，反向竞价比例 = （ 当前时间 - 文章发表时间 ） / 30分钟

（4）综合以上，当该赞在点赞次序中最终排位靠后，点赞收益是受损的，将低于全站的平均点赞收益，同样如果在文章发表后30分钟内点赞，也会低于预期收益（差额给作者）。这就意味着，对文章进行点赞，既不能太早也不能太晚。

总结：结合文章奖励分配和点赞奖励分配，同一篇文章的点赞者先全部作为一个整体，参与全站的收益分配竞争；获得的收益再在内部进行分配。注意，外部竞争和内部分配，都是非线性的。s的取值，是系统用于调整激励以实现系统目标的主要参数。

- **点赞者贡献（由点赞者PGC决定）**
- **点赞时间（30分钟以内有惩罚，30分钟以后为1）**
- **位置因子（越靠前收益越大）**
